# Sorting the Transcriptomic Dataset for Mus Musculus by Various Conditional Expression Levels  

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df2 = pd.read_csv("transcriptomedata.csv")
#for col in df2.columns: 
#    print(col) 
```

# Transcripts with high enrichment profiles in oocytes and the heart: sorted into subgroups by deviations from the median. 

```python
df2_3XM = df2[df2["oocyte"]>(3*df2["oocyte"].median())]
df2_10XM = df2[df2["oocyte"]>(10*df2["oocyte"].median())]
df2_3XM_ooc_3XM_heart = df2_3XM[df2_3XM["heart"]>(3*df2_3XM["heart"].median())]
df2_3XM_ooc_10XM_heart = df2_3XM[df2_3XM["heart"]>(10*df2_3XM["heart"].median())]
df2_10XM_ooc_3XM_heart = df2_10XM[df2_10XM["heart"]>(3*df2_10XM["heart"].median())]
df2_10XM_ooc_10XM_heart = df2_10XM[df2_10XM["heart"]>(10*df2_10XM["heart"].median())]
```

```python
df2_3O_3H_by_max_ooc = df2_3XM_ooc_3XM_heart.sort_values(by=["oocyte"], ascending = False)
df2_3O_3H_by_max_heart = df2_3XM_ooc_3XM_heart.sort_values(by=["heart"], ascending = False)
df2_3XM_ooc_3XM_heart.shape

import csv
with open('df2_3XM_ooc_3XM_heart.csv', 'w', newline='') as csvfile:
    table = csv.writer(csvfile, delimiter=',',
                            quotechar='|', quoting=csv.QUOTE_MINIMAL)
    for row in df2_3XM_ooc_3XM_heart:
        table.writerow(df2_3XM_ooc_3XM_heart[row]) 
```

```python
df2_3O_10H_by_max_ooc = df2_3XM_ooc_10XM_heart.sort_values(by=["oocyte"], ascending = False)
df2_3O_10H_by_max_heart = df2_3XM_ooc_10XM_heart.sort_values(by=["heart"], ascending = False)
df2_3XM_ooc_10XM_heart.shape

import csv
with open('df2_3XM_ooc_10XM_heart.csv', 'w', newline='') as csvfile:
    table = csv.writer(csvfile, delimiter=',',
                            quotechar='|', quoting=csv.QUOTE_MINIMAL)
    for row in df2_3XM_ooc_10XM_heart:
        table.writerow(df2_3XM_ooc_10XM_heart[row]) 
```

```python
df2_10O_3H_by_max_ooc = df2_10XM_ooc_3XM_heart.sort_values(by=["oocyte"], ascending = False)
df2_10O_3H_by_max_heart = df2_10XM_ooc_3XM_heart.sort_values(by=["heart"], ascending = False)
df2_10XM_ooc_3XM_heart.shape

import csv
with open('df2_10XM_ooc_3XM_heart.csv', 'w', newline='') as csvfile:
    table = csv.writer(csvfile, delimiter=',',
                            quotechar='|', quoting=csv.QUOTE_MINIMAL)
    for row in df2_10XM_ooc_3XM_heart:
        table.writerow(df2_10XM_ooc_3XM_heart[row]) 
```

```python
df2_10O_10H_by_max_ooc = df2_10XM_ooc_10XM_heart.sort_values(by=["oocyte"], ascending = False)
df2_10O_10H_by_max_heart = df2_10XM_ooc_10XM_heart.sort_values(by=["heart"], ascending = False)
df2_10XM_ooc_10XM_heart.shape

import csv
with open('df2_10XM_ooc_10XM_heart.csv', 'w', newline='') as csvfile:
    table = csv.writer(csvfile, delimiter=',',
                            quotechar='|', quoting=csv.QUOTE_MINIMAL)
    for row in df2_10XM_ooc_10XM_heart:
        table.writerow(df2_10XM_ooc_10XM_heart[row]) 
```
